```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
def analyze_dataframe(df):
    summary = []

    for col in df.columns:
        col_data = df[col]
        info = {
            "column": col,
            "dtype": col_data.dtype,
            "missing": col_data.isna().sum(),
            "missing_%": round(col_data.isna().mean() * 100, 2),
            "unique": col_data.nunique()
        }

        if pd.api.types.is_numeric_dtype(col_data):
            info.update({
                "mean": col_data.mean(),
                "std": col_data.std(),
                "min": col_data.min(),
                "max": col_data.max()
            })
        else:
            info.update({
                "top_value": col_data.mode().iloc[0] if not col_data.mode().empty else None,
                "top_count": col_data.value_counts().iloc[0] if not col_data.value_counts().empty else None
            })

        summary.append(info)

    return pd.DataFrame(summary)
```

```python
analysis = analyze_dataframe(books)
analysis
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] =2026- books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style = "white")
plt.figure(figsize = (8, 6))
heatmap = sns.heatmap(correlation_matrix, annot = True, fmt = ".2f", cmap = "coolwarm",
                      cbar_kws = {"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
book_missing = books[~books["description"].isna() & 
      ~books["num_pages"].isna() &
      ~books["average_rating"].isna() &
      ~books["published_year"].isna() 
      ]
```

```python
import matplotlib.pyplot as plt

counts = book_missing["categories"].value_counts()

N = 20
top = counts.head(N)
others = counts.iloc[N:].sum()

plot_data = top._append(
    pd.Series({"Other": others})
)

plt.figure(figsize=(10, 4))
plt.bar(plot_data.index, plot_data.values)
plt.xticks(rotation=45, ha="right")
plt.title("Top categories (others grouped)")
plt.tight_layout()
plt.show()
```
